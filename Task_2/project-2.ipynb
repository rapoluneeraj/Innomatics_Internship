```python
print(df.Gender.value_counts())
print(df.Gender.unique())
print(df.Gender.max()==df.Gender)
print(df.DOL[df.Gender.max()==df.Gender].value_counts())
print(df.DOL[df.Gender.min()==df.Gender].value_counts())
##2.Most number of candidates recruited are males and candidates leaving the organisation are more in males compared to females
```

```python
print(df.Designation.value_counts())
df.Designation[(df.CollegeTier==1)&(df.Designation=='software engineer')].value_counts()
##No of candidates from tier 1 college choosing the most chosen designation i.e is 50
```

```python
a=df.Quant.median()
print(a)
b=df.Designation[(df.Quant>a)&(df.Gender=='f')].value_counts()
c=df.Designation[(df.Quant>a)&(df.Gender=='m')].value_counts()
print(b)
print(c)
##Males and females with quant score more than average candidates preffered software engineer as profession
```

```python
df.JobCity[(df['10board']=='cbse')&(df.Gender=='f')].value_counts()
##Noida is the highest jobcity with females from cbse board getting recruited
```

```python
print(df.Gender[df.CivilEngg>-1].value_counts())
print(df.Gender[df.CivilEngg.max()])
```

```python
df.groupby('Gender').agg({'10percentage': ['min', 'max','median'],'12percentage': ['min', 'max','median'], 'Salary': 'median'})
##Despite females dominating males with more average score in 10th and 12 th boards, yet both genders earn equally
```

```python
df.groupby('Degree').agg({'10percentage': ['mean','median'],'12percentage': ['mean'
                                                                             ,'median'], 'Salary':[ 'median','mean']})

##Despite the average marks obtained by M.Sc(Tech) students in 12th board being 55% the candidates managed
##to get average of 3,20,000 Lakh, which is more compared to MCA and B.Tech Students 
```

```python
df.groupby('Degree').agg({ 'English': ['max','min','median'],'Quant': ['max','min','median'],'Logical': ['max','min','median']})
##M.Sc(Tech) students scored more minimum marks in logical,english,Quant than other degree students

```
